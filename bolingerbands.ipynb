```python
import numpy as np
import pandas as pd
import pandas_datareader as web
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = web.DataReader('DOGUB.IS',data_source='yahoo',start='2018-01-01',end='2020-12-24')
data.shape
data
```

```python
period = 20
```

```python
data['SMA'] = data['Close'].rolling(window = period).mean()
data['STD'] = data['Close'].rolling(window = period).std()
```

```python
data['Yukarı'] = data['SMA'] + (data['STD']*2)
data['Aşağı'] = data['SMA'] - (data['STD']*2)
```

```python
sutun_liste = ['Close','SMA','Aşağı','Yukarı']
```

```python
data[sutun_liste].plot(figsize=(25,12))
plt.title('PEKGY - Bollinger Bands Stratejisi')
plt.ylabel('Türk Lirası')
plt.show()
```

```python
data
```

```python
yeni_veri = data[period-1:]
```

```python
yeni_veri
```

```python
def sinyal_al(yeniveriler):
    al_sinyal = []
    sat_sinyal=[]
    for i in range(len(yeniveriler['Close'])):
        if yeniveriler['Close'][i] < yeniveriler['Aşağı'][i]:
            al_sinyal.append(yeniveriler['Close'][i])
            sat_sinyal.append(np.nan)
        elif yeniveriler['Close'][i] > yeniveriler['Yukarı'][i]:
            al_sinyal.append(np.nan)
            sat_sinyal.append(yeniveriler['Close'][i])
        else: 
            al_sinyal.append(np.nan)
            sat_sinyal.append(np.nan)
    return(al_sinyal,sat_sinyal)
```

```python
yeni_veri['Al'] = sinyal_al(yeni_veri)[0]
yeni_veri
```

```python
fig = plt.figure(figsize=(25,12))
ax = fig.add_subplot(1,1,1)
x_axis = yeni_veri.index
ax.fill_between(x_axis,yeni_veri['Yukarı'],yeni_veri['Aşağı'],color='lightgray')
ax.plot(x_axis,yeni_veri['SMA'],color='red',label='Basit Hareketli Ortalama',alpha=0.5)
ax.plot(x_axis,yeni_veri['Close'],color='blue',label='Kapanış Fiyatı',alpha=0.5)
ax.scatter(x_axis,yeni_veri['Al'],color='darkgreen',label='Al',alpha=0.9)
ax.set_title('SMART Bolinger Bands')
ax.set_ylabel('Türk Lirası')
ax.set_xlabel('Tarih')
plt.xticks(rotation = 45)
ax.legend()
plt.show()
```
